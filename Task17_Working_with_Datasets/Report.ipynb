```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
# Compare the average income based on ethnicity
income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print(income_by_ethnicity)
```

```python
# On average, do married or single people have a higher balance?
balance_by_marital_status = df.groupby('Married')['Balance'].mean()


if balance_by_marital_status['Yes'] > balance_by_marital_status['No']:
    print(f"On average, married people have a higher balance of {balance_by_marital_status['Yes']:.2f}.")
else:
    print(f"On average, unmarried people have a higher balance of {balance_by_marital_status['No']:.2f}.")

print(balance_by_marital_status)
```

```python
# How many females do we have information for vs how many males?

# Use value_counts() to get counts for each gender
gender_counts = df['Gender'].value_counts()

# Print gender counts with their frequencies
print(gender_counts)
```

```python
# What is the highest income in our dataset?
highest_income = df['Income'].max()
print (highest_income)
```

```python
# What is the lowest income in our dataset?
lowest_income = df['Income'].min()
print (lowest_income)
```

```python
# How many cards do we have recorded in our dataset?
total_cards_recorded = df['Cards'].sum()
print (total_cards_recorded)
```
