```python
# Importing necessary libraries
import pandas as pd                 # Used for data manipulation and analysis.
import matplotlib.pyplot as plt     # Used for creating static, animated, and interactive visualizations in Python.
from sklearn.linear_model import LinearRegression   # Used for performing linear regression modeling.
import seaborn as sns
from scipy.stats import linregress

```

```python
# Importing the dataset
df = pd.read_csv('insurance.csv')   # Reading a CSV file and creating a pandas DataFrame object.
df.head()                           # Displaying the first 5 rows of the dataframe to ensure it was properly loaded.
```

```python
"""
Thie below code creates a scatter plot between two variables, age and charges, from the pandas DataFrame `df` using the `plt.scatter()` function in Matplotlib.

The first line of code uses `plt.scatter()` to create a scatter plot with age on the x-axis and charges on the y-axis. 
This plots each value of age and its corresponding charge as a point on the scatter plot.

The second line assigns a label to the x-axis using `plt.xlabel()`, which sets the text for the x-axis label to "Age", 
and the third line assigns a label to the y-axis using `plt.ylabel()`, which sets the text for the y-axis label to "Charges".

The fourth line gives a title to the scatter plot using `plt.title()`, which sets the title of the plot to "Age vs Charges".

Finally, the fifth line calls `plt.show()` to display the plot.

Overall, this code generates a simple scatter plot visualizing the relationship between age and charges in `df`
"""
```

```python
# Creating a scatter plot between age and charges from the DataFrame df
plt.scatter(df['age'], df['charges'])

# Assigning labels for the x and y-axis
plt.xlabel('Age')
plt.ylabel('Charges')

# Giving a title to the plot
plt.title('Age vs Charges')

# Displaying the plot.
plt.show()

```

```python
"""
Thie below code uses the scikit-learn library's LinearRegression class to perform linear regression on a dataset containing 
information about healthcare costs. Specifically, it selects one feature ('age') from a Pandas DataFrame and stores it in 
the variables X, and another feature ('charges') in variable y. It then creates an instance of the LinearRegression class 
and fits the model with the X and y variables. The fitted model is used to generate predictions, which are stored in the 
'predictions' variable.

The code then prints the intercept and coefficient of determination (R^2) of the fitted linear regression model. 
The intercept is the value of y when x is zero, and the coefficient of determination indicates how well the linear 
regression model fits the data.

Finally, the code plots actual charges vs predicted charges on a scatter plot using the matplotlib library. 
The plot is labeled and titled and includes additional information about the model. The plt.text() method is used 
to display this information on the graph, including the intercept, slope, and R-squared values. 

Overall, this code demonstrates how to use scikit-learn and matplotlib libraries for data modeling and visualization.
"""
```

```python

# Selecting the 'age' column from DataFrame df and storing it in X
X = df[['age']]

# Storing the 'charges' column from DataFrame df in y
y = df['charges']

# Creating a Linear Regression object for the model
model = LinearRegression()

# Fitting the model with X and y variables
model.fit(X, y)

# Using the fitted model to generate predictions based on the input values of X (age)
predictions = model.predict(X)

# Printing the intercept and coefficient of determination (R^2) of the fitted linear regression model
print('Intercept:', model.intercept_)
print('Coefficient of determination (R^2):', model.score(X, y))

# Plotting actual charges vs predicted charges
plt.scatter(y, predictions)
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual Charges vs Predicted Charges")

# Displaying additional information about the model
plt.text(20000, 5000,
         f"Intercept: {model.intercept_:.2f}\n"
         f"Slope: {model.coef_[0]:.2f}\n"
         f"R-squared: {model.score(X, y):.4f}")
plt.show()
```

```python


# Creating subplots to display line plot and histogram
fig, axs = plt.subplots(ncols=2, figsize=(12, 5))

# Plotting charges vs age relationship using a line plot
sns.lineplot(data=df, x='age', y='charges', ax=axs[0], color='red')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Insurance Charges')
axs[0].set_title('Charges vs Age Relationship')


# Plotting histogram of age distribution in the dataset
sns.histplot(data=df, x='age', ax=axs[1], color='blue')
axs[1].set_xlabel('Age')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Age Distribution')



# Adjusting the space between subplots and bottom margin
plt.subplots_adjust(wspace=0.3, bottom=0.1)

# Displaying the plots
plt.show()


# Create a scatter plot with regression line for the age vs charges relationship
x = df['age']
y = df['charges']
slope, intercept, r_value, p_value, std_err = linregress(x, y)

plt.scatter(x, y, alpha=0.5)
plt.plot(x, intercept + slope * x, 'r', label='fitted line')
plt.legend()
plt.title('Scatter Plot of Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Charges')



# Create a hexbin plot for the age vs charges relationship
plt.hexbin(x='age', y='charges', data=df, cmap='YlOrRd', gridsize=20)
plt.colorbar()
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Hexbin Plot of Charges vs Age')



# Create a jointplot for the age vs charges relationship
sns.jointplot(x='age', y='charges', data=df, kind='scatter')
plt.suptitle('Joint Plot of Charges vs Age', y=1.05)



# Create a scatter plot with age on the x-axis and charges on the y-axis
plt.scatter(df['age'], df['charges'], alpha=0.5)
plt.title('Scatter Plot of Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Charges')



# Display all plots
plt.show()
```
